# 2. 애플주가 분석

```python
import yfinance as yf
import matplotlib.pyplot as plt

# 애플 주가 데이터를 가져오기
ticker = 'AAPL'
apple_stock = yf.Ticker(ticker)
hist = apple_stock.history(period="1y")  # 최근 1년간의 데이터

# 차트 그리기
plt.figure(figsize=(10, 5))
plt.plot(hist.index, hist['Close'], label='Close Price')
plt.title('Apple Stock Price (Last 1 Year)')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

# 3. 이미지 저장

```python
import yfinance as yf
import matplotlib.pyplot as plt

# 애플 주가 데이터를 가져오기
ticker = 'AAPL'
apple_stock = yf.Ticker(ticker)
hist = apple_stock.history(period="1y")  # 최근 1년간의 데이터

# 차트 그리기
plt.figure(figsize=(10, 5))
plt.plot(hist.index, hist['Close'], label='Close Price')
plt.title('Apple Stock Price (Last 1 Year)')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.legend()
plt.grid(True)

# 차트를 이미지 파일로 저장
plt.savefig('apple_stock_price.png')

# 차트 보여주기
plt.show()
```
